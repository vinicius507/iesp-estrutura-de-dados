## Lista Sequencial

```python
import numpy as np
import numpy.typing as npt
from collections.abc import MutableSequence

class ListaSequencial:
    def __init__(self, capacidade: int):
        self.capacidade: int = capacidade
        self.valores: npt.NDArray[np.generic[int]] = np.empty(self.capacidade, dtype=int)
        self.ultima_posicao: int = -1

    def insere(self, valor: int):
        if self.ultima_posicao == self.capacidade - 1:
            print('Capacidade máxima atingida')
        else:
            self.ultima_posicao += 1
            self.valores[self.ultima_posicao] = valor

    def pesquisar(self, valor: int):
        for i in range(0, self.ultima_posicao + 1):
            if self.valores[i] == valor:
                return i
        return -1

    def excluir(self, valor: int):
        posicao = self.pesquisar(valor)
        if posicao == -1:
            return -1
        else:
            for i in range(posicao, self.ultima_posicao):
                self.valores[i] = self.valores[i + 1]
            self.ultima_posicao -= 1

    def imprime(self):
        if self.ultima_posicao == -1:
            print("O vetor está vazio")
        else:
            for i in range(self.ultima_posicao + 1):
                print(f"{i} - {self.valores[i]}")
```

## Ordenação

```python
def bubble_sort(lista: MutableSequence[int]) -> MutableSequence[int]:
    n = len(lista)

    for i in range(n - 1):
        for j in range(n - i - 1):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
    return lista

def selection_sort(lista: MutableSequence[int]) -> MutableSequence[int]:
    n = len(lista)

    for i in range(n):
        smallest_idx = i
        for j in range(i, n):
            if lista[j] < lista[smallest_idx]:
                smallest_idx = j
        if smallest_idx != i:
            lista[i], lista[smallest_idx] = lista[smallest_idx], lista[i]
    return lista


def choose_pivot(arr: MutableSequence[int], start: int, end: int) -> tuple[int, int]:
    idx = (end + start) // 2

    return arr[idx], idx

def swap(arr: MutableSequence[int], i: int, j: int):
    arr[i], arr[j] = arr[j], arr[i]

def partition(arr: MutableSequence[int], start: int, end: int) -> int:
    i = start - 1
    pivot, pivot_idx = choose_pivot(arr, start, end)

    swap(arr, pivot_idx, end)
    for j in range(start, end):
        if arr[j] < pivot:
            i += 1
            swap(arr, i, j)
    i += 1
    swap(arr, i, end)
    return i

def quick_sort(arr: MutableSequence[int], start: int | None = None, end: int | None = None):
    if start is None:
        start = 0
    if end is None:
        end = len(arr) - 1

    if start >= end:
        return
    pi = partition(arr, start, end)
    quick_sort(arr, start, pi - 1)
    quick_sort(arr, pi + 1, end)
```

## Fila

```python
class Fila:
    def __init__(self, capacidade: int):
        self.capacidade: int = capacidade
        self.inicio: int = 0
        self.final: int = -1
        self.numero_elementos: int = 0
        self.valores: npt.NDArray[np.int_]  = np.empty(self.capacidade, dtype=np.int_)

    def fila_vazia(self):
        return self.numero_elementos == 0

    def fila_cheia(self):
        return self.numero_elementos == self.capacidade

    def enfileirar(self, valor: int):
        if self.fila_cheia():
            print('A fila está cheia')
            return
        if self.final == self.capacidade - 1:
            self.final = -1
        self.final += 1
        self.valores[self.final] = valor
        self.numero_elementos += 1

    def desenfileirar(self):
        if self.fila_vazia():
            print('A fila já está vazia')
            return
        temp = self.valores.item(self.inicio)
        self.inicio += 1
        if self.inicio == self.capacidade - 1:
            self.inicio = 0
        else:
            self.numero_elementos -= 1
            return temp

    def primeiro(self):
        if self.fila_vazia():
            return -1
        else:
            return self.valores.item(self.inicio)

    def visualizar(self):
      print('Tamanho da Fila..: ', self.capacidade)
      print('Início da fila...: ', self.inicio)
      print('Final da fila....: ', self.final)
      print('Variável Nu. El..:', self.numero_elementos)

    def listageral(self):
      for i in range(self.capacidade):
        print(self.valores.item(i), end=' ')
```

## Pilha

```python
class Pilha:

    def __init__(self):
        self.items: list[int] = []

    def empilhar(self, item: int):
        self.items = self.items + [item]

    def desempilhar(self):
        if not self.is_vazia():
            item_removido = self.items[-1]
            self.items = self.items[:-1]
            return item_removido
        else:
            print("A pilha está vazia. Não é possível desempilhar.")

    def topo(self):
        if not self.is_vazia():
            return self.items[-1]
        else:
            print("A pilha está vazia. Não há topo para visualizar.")

    def is_vazia(self):
        return len(self.items) == 0

    def tamanho(self):
        return len(self.items)

    def imprimir(self):
        if not self.is_vazia():
            print("Itens da pilha:")
            for item in self.items:
                print(item)
        else:
            print("A pilha está vazia. Não há itens para imprimir.")
```

04 - Após carregar a base de dados, qual dos métodos de classifição (bubble_sort, select_sort e quick_sort) é o mais eficiente?


```python
from typing import Callable
from functools import wraps

def metrificar(fn: Callable[[MutableSequence[int]], MutableSequence[int] | None]):
    @wraps(fn)
    def wrap(lista: MutableSequence[int]):
        inicio = time.time()
        _ = fn(lista) # Não interessa agora
        fim = time.time()
        return fim - inicio
    return wrap

def questao_04():
    bsort = metrificar(bubble_sort)
    ssort = metrificar(selection_sort)
    qsort = metrificar(quick_sort)
    populacao_testes = list(range(200))
    testes = [random.sample(populacao_testes, i) for i in (1, 5, 10, 100, 200)]

    for teste in testes:
        winner = "Ninguém"
        maior_velocidade = float('inf')
        bsorted, ssorted, qsorted = bsort(teste), ssort(teste), qsort(teste)

        if bsorted < maior_velocidade:
            winner = 'Bubble Sort'
            maior_velocidade = bsorted
        if ssorted < maior_velocidade:
            winner = 'Selection Sort'
            maior_velocidade = ssorted
        if qsorted < maior_velocidade:
            winner = 'Quick Sort'
        print(f"{teste=}\n{winner=} tempos=bsort={bsorted:.6f}, ssort={ssorted:.6f}, qsort={qsorted:.6f}\n")

questao_04()
```
