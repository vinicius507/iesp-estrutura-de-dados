## Círculo

Crie uma classe chamada “Circulo” que tenha um atributo “raio”. Implemente um método chamado “calcular_area” que retorna a área do círculo.

```python
import math

class Circulo:
    def __init__(self, raio: int):
        self.raio: int = raio

    def calcular_area(self):
        return 2 * math.pi * self.raio

circulo = Circulo(10)
circulo.calcular_area()
```

## Livro

Crie uma classe chamada “Livro” que tenha atributos “titulo” e “autor”. Implemente um método chamado “detalhes” que retorna uma string com as informações do livro.


```python
class Livro:
    def __init__(self, titulo: str, autor: str) -> None:
        self.titulo: str = titulo
        self.autor: str = autor

    def detalhes(self) -> str:
        return f"{self.titulo} por {self.autor}"

livro = Livro("O Programador Pragmático", "Andrew Hunt")
livro.detalhes()
```

## Retangulo

Crie uma classe chamada “Retangulo” que tenha atributos “base” e “altura”. Implemente um método chamado “calcular_area” que retorna a área do retângulo.

```python
class Retangulo:
    def __init__(self, base: int, altura: int) -> None:
        self.base: int = base
        self.altura: int = altura

    def calcular_area(self) -> int:
        return self.base * self.altura

retangulo = Retangulo(10, 5)
retangulo.calcular_area()
```

## Conta Bancária

Crie uma classe chamada “ContaBancaria” que tenha atributos “saldo” e “titular”. Implemente métodos “depositar” e “sacar” para manipular o saldo.

```python
from decimal import Decimal

class SaldoInsuficienteError(Exception):
    def __init__(self, saldo_disponivel: Decimal):
        super().__init__(f"o saldo da conta é insuficiente. saldo disponível: {saldo_disponivel}")

class ContaBancaria:
    def __init__(self, titular: str) -> None:
        self.titular: str = titular
        self.saldo: Decimal = Decimal()

    def depositar(self, valor: Decimal):
        self.saldo += valor

    def sacar(self, valor: Decimal):
        if valor > self.saldo:
            raise SaldoInsuficienteError(self.saldo)
        self.saldo -= valor

conta = ContaBancaria("Vinicius")
conta.depositar(Decimal(10))
print(conta.saldo)
conta.depositar(Decimal(10))
print(conta.saldo)
conta.sacar(Decimal(19))
print(conta.saldo)
try:
    conta.sacar(Decimal(100))
except SaldoInsuficienteError as e:
    print(e)
```

## Pessoa

Crie uma classe chamada “Pessoa” com atributos “nome” e “idade”. Implemente um método chamado “falar” que imprime uma mensagem com o nome da pessoa.

```python
class Pessoa:
    def __init__(self, nome: str, idade: int) -> None:
        self.nome: str = nome
        self.idade: int = idade

    def falar(self):
        print(f"Olá, meu nome é {self.nome} e tenho {self.idade} anos.")

vini = Pessoa("Vini", 27)
camylla = Pessoa("Camylla", 28)

vini.falar()
camylla.falar()
```

## Produto

Crie uma classe chamada “Produto” com atributos “nome”, “preço” e “quantidade”. Implemente um método chamado “calcular_total” que retorna o valor total do produto (preço * quantidade).

```python
from decimal import Decimal

class Produto:
    def __init__(self, nome: str, preco: Decimal, quantidade: int) -> None:
        self.nome: str = nome
        self.preco: Decimal = preco
        self.quantidade: int = quantidade

    def calcular_total(self) -> Decimal:
        return self.preco * self.quantidade

produto = Produto("Caixa de Som", Decimal("120"), 5)
produto.calcular_total()
```

## Carro

Crie uma classe chamada “Carro” com atributos “marca”, “modelo” e “ano”. Implemente um método chamado “detalhes” que retorna uma string com as informações do carro.

```python
class Carro:
    def __init__(self, marca: str, modelo: str, ano: int) -> None:
        self.marca: str = marca
        self.modelo: str = modelo
        self.ano: int = ano

    def detalhes(self) -> str:
        return f"{self.marca} {self.modelo} ({self.ano})"

carro = Carro("Toyota", "Corolla", 2020)
carro.detalhes()
```

## Aluno

Crie uma classe chamada “Aluno” com atributos “nome” e “notas”. Implemente um método chamado “calcular_media” que retorna a média das notas do aluno.

```python
from decimal import Decimal

class Aluno:
    def __init__(self, nome: str, notas: list[Decimal]) -> None:
        self.nome: str = nome
        self.notas: list[Decimal] = notas

    def calcular_media(self) -> Decimal:
        return sum(self.notas) / Decimal(len(self.notas))

aluno = Aluno("Vinicius", [Decimal("10"), Decimal("2.5")])
aluno.calcular_media()
```

## Triângulo

Crie uma classe chamada “Triangulo” com atributos “lado1”, “lado2” e “lado3”. Implemente um método chamado “calcular_perimetro” que retorna o perímetro do triângulo.

```python
class Triangulo:
    def __init__(self, lado1: int, lado2: int, lado3: int) -> None:
        self.lado1: int = lado1
        self.lado2: int = lado2
        self.lado3: int = lado3

    def calcular_perimetro(self) -> int:
        return self.lado1 + self.lado2 + self.lado3

triangulo = Triangulo(3, 3, 3)
triangulo.calcular_perimetro()
```

## Funcionário

Crie uma classe chamada “Funcionario” com atributos “nome”, “salario” e “cargo”. Implemente um método chamado “aumentar_salario” que recebe um valor percentual de aumento e atualiza o salário do funcionário.

```python
from decimal import Decimal

class Funcionario:
    def __init__(self, nome: str, salario: Decimal, cargo: str) -> None:
        self.nome: str = nome
        self.salario: Decimal = salario
        self.cargo: str = cargo

    def aumentar_salario(self, percentual: Decimal):
        self.salario += self.salario * percentual / Decimal("100")

funcionario = Funcionario("Vinicius", Decimal(1000.0), "Desenvolvedor")
funcionario.aumentar_salario(Decimal("10"))
print(funcionario.salario)
```
