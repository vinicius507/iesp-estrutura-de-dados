# Lista 003: Lista Sequencial

```python
import numpy as np

class ListaSequencial:
    def __init__(self, capacidade: int):
        self.capacidade: int = capacidade
        self.valores: np.ndarray = np.empty(self.capacidade, dtype=int)
        self.ultima_posicao: int = -1

    def insere(self, valor: int):
        if self.ultima_posicao == self.capacidade - 1:
            print('Capacidade máxima atingida')
        else:
            self.ultima_posicao += 1
            self.valores[self.ultima_posicao] = valor

    def pesquisar(self, valor: int):
        for i in range(0, self.ultima_posicao + 1):
            if self.valores[i] == valor:
                return i
        return -1

    def excluir(self, valor: int):
        posicao = self.pesquisar(valor)
        if posicao == -1:
            return -1
        else:
            for i in range(posicao, self.ultima_posicao):
                self.valores[i] = self.valores[i + 1]
            self.ultima_posicao -= 1

    def imprime(self):
        if self.ultima_posicao == -1:
            print("O vetor está vazio")
        else:
            for i in range(self.ultima_posicao + 1):
                print(f"{i} - {self.valores[i]}")
```

1. Crie uma lista sequencial com capacidade para 5 elementos. Insira as matrículas dos seguintes alunos: 101, 102, 103, 104. Depois, imprima a lista para verificar os valores armazenados.

```python
lista = ListaSequencial(5)
lista.insere(101)
lista.insere(102)
lista.insere(103)
lista.insere(104)
lista.imprime()
```

```python
lista = ListaSequencial(5)
lista.insere(45)
lista.insere(33)
lista.insere(27)
lista.insere(89)
lista.insere(56)
lista.imprime()
lista.insere(72)
```

3. Considere que foram cadastrados os clientes com os códigos 11, 22, 33, 44, 55. Pesquise se o cliente com código 33 está presente e informe sua posição na lista.

```python
lista = ListaSequencial(5)
lista.insere(11)
lista.insere(22)
lista.insere(33)
lista.insere(44)
lista.insere(55)
posicao = lista.pesquisar(33)
if posicao != -1:
    print(f"O cliente com código 33 está na posição {posicao}")
else:
    print("O cliente com código 33 não está na lista")
```
