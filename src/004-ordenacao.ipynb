1. Escreva uma função em Python para ordenar um vetor de inteiros em ordem crescente usando o algoritmo de Select Sort.

```python
def selection_sort(lista: list[int]) -> list[int]:
    n = len(lista)

    for i in range(n):
        idx_menor = i

        for j in range(i, n):
            if lista[j] < lista[idx_menor]:
                idx_menor = j
        if idx_menor != i:
            lista[i], lista[idx_menor] = lista[idx_menor], lista[i]
    return lista

print(selection_sort([1, 2, 3]))
print(selection_sort([3, 2, 1]))
print(selection_sort(random.sample(range(-10, 10), 10)))
```

2. Escreva uma função em Python para ordenar um vetor de inteiros, ele deve receber um parâmetro que serve como chave para realizar a ordenação crescente ou decrescente.

```python
def ordenar(lista: list[int], *, crescente: bool = True) -> list[int]:
    n = len(lista)

    def comparar(a: int, b: int, crescente: bool) -> bool:
        if crescente:
            return a < b
        return a > b

    def trocar(i: int, j: int) -> None:
        lista[i], lista[j] = lista[j], lista[i]

    for i in range(n):
        idx_alvo = i

        for j in range(i, n):
            if comparar(lista[j], lista[idx_alvo], crescente):
                idx_alvo = j
        if idx_alvo != i:
            trocar(i, idx_alvo)
    return lista

print(ordenar([1, 2, 3]))
print(ordenar([3, 2, 1]))
print(ordenar(random.sample(range(-10, 10), 10)))
print(ordenar([1, 2, 3], crescente=False))
print(ordenar([3, 2, 1], crescente=False))
print(ordenar(random.sample(range(-10, 10), 10), crescente=False))
```

3. Escreva um programa que encontre o elemento de maior valor em um vetor de inteiros não ordenado sem usar a função `max()`. Em seguida, encontre o elemento mínimo sem usar a função `min()`.

```python
def maximo(lista: list[int]) -> int:
    maior = lista[0]

    for num in lista:
        if num > maior:
            maior = num
    return maior


def minimo(lista: list[int]) -> int:
    menor = lista[0]

    for num in lista:
        if num < menor:
            menor = num
    return menor

testes = [
    [1, 2, 3],
    [3, 2, 1],
    random.sample(range(-10, 10), 10)
]

for teste in testes:
    print(f'Lista: {teste}')
    print(f'Máximo: {maximo(teste)}')
    print(f'Mínimo: {minimo(teste)}')
```

4. Crie uma função que recebe um vetor de números inteiros e retorna o segundo menor número. Certifique-se de que sua função funcione mesmo se houver números duplicados no vetor.

```python
def segundo_menor(lista: list[int]) -> int:
    if len(lista) < 2:
        raise ValueError("A lista deve conter pelo menos dois elementos.")

    menor, segundo_menor = float('inf'), float('inf')

    for num in lista:
        if num < menor:
            segundo_menor = menor
            menor = num
            continue
        if menor < num < segundo_menor:
            segundo_menor = num
    if isinstance(segundo_menor, float):
        raise ValueError("Não há um segundo menor número distinto na lista.")
    return segundo_menor

testes = [
    [1, 2, 3],
    [3, 2, 1],
    [1, 1, 2, 3],
    random.sample(range(-10, 10), 10)
]

for teste in testes:
    print(f'Lista: {teste}')
    print(f'Segundo menor: {segundo_menor(teste)}')
```

5. Implemente uma função que aceite um vetor de números inteiros e remova todos os elementos duplicados, retornando o vetor resultante sem duplicatas. Não é permitido utilizar a função “set()”

```python
def remover_duplicatas(lista: list[int]) -> list[int]:
    resultado: list[int] = []

    for num in lista:
        if num not in resultado:
            resultado.append(num)
    return resultado

testes = [
    [1, 2, 3],
    [1, 1, 2, 3],
    [3, 2, 1],
    [4, 10, 4, 5, 1, 1, 6, 20, 30, 6, 6, 4, 10],
    random.sample(range(-10, 10), 10) + random.sample(range(-10, 10), 10)
]

for teste in testes:
    print(f'Lista original: {teste}')
    print(f'Lista sem duplicatas: {remover_duplicatas(teste)}')
```

6. Escreva um programa que ordene um vetor de inteiros em ordem decrescente e, em seguida, conte quantos números pares e quantos números ímpares existem no vetor ordenado.

```python
def ordenar_e_contar(lista: list[int]) -> tuple[list[int], int, int]:
    pares = 0
    n = len(lista)

    for i in range(n):
        idx_maior = i

        for j in range(i, n):
            if lista[j] > lista[idx_maior]:
                idx_maior = j
        if idx_maior != i:
            lista[i], lista[idx_maior] = lista[idx_maior], lista[i]
        if lista[i] % 2 == 0:
            pares += 1
    impares = n - pares
    return lista, pares, impares

testes = [
    [1, 2, 3],
    [3, 2, 1],
    random.sample(range(-10, 10), 10),
    random.sample(range(-15, 15), 10),
    random.sample(range(-20, 20), 10)
]

for teste in testes:
    ordenada, pares, impares = ordenar_e_contar(teste)
    print(f'Lista ordenada: {ordenada}')
    print(f'Número de pares: {pares}')
    print(f'Número de ímpares: {impares}')
```

7. Crie uma função que aceite um vetor de números inteiros e retorne o terceiro maior número. Certifique-se de que sua função funcione mesmo se houver números duplicados no vetor.

```python
def terceiro_maior(lista: list[int]) -> int:
    if len(lista) < 3:
        raise ValueError("A lista deve conter pelo menos três elementos.")

    maior, segundo_maior, terceiro_maior = float('-inf'), float('-inf'), float('-inf')

    for num in lista:
        if num > maior:
            terceiro_maior = segundo_maior
            segundo_maior = maior
            maior = num
            continue
        if maior > num > segundo_maior:
            terceiro_maior = segundo_maior
            segundo_maior = num
            continue
        if segundo_maior > num > terceiro_maior:
            terceiro_maior = num

    if isinstance(terceiro_maior, float):
        raise ValueError("Não há um terceiro maior número distinto na lista.")
    return terceiro_maior

testes = [
    [1, 2, 3],
    [3, 2, 1],
    [1, 1, 2, 3],
    random.sample(range(-10, 10), 10)
]

for teste in testes:
    print(f'Lista: {teste}')
    print(f'Terceiro maior: {terceiro_maior(teste)}')
```

8. Crie uma função que receba um vetor de números inteiros e retorne a mediana, ou seja, o valor do meio quando o vetor é ordenado. Certifique-se de que sua função funcione para vetores com quantidade de elementos ímpares.

```python
def mediana(lista: list[int]) -> float:
    n = len(lista)
    lista_copia = lista.copy()

    if n == 0:
        raise ValueError("A lista não pode ser vazia.")

    def particionar(inicio: int, fim: int) -> int:
        pivot = lista_copia[fim]
        i = inicio

        for j in range(inicio, fim):
            if lista_copia[j] <= pivot:
                lista_copia[i], lista_copia[j] = lista_copia[j], lista_copia[i]
                i += 1
        lista_copia[i], lista_copia[fim] = lista_copia[fim], lista_copia[i]
        return i

    def k_menor(inicio: int, fim: int, k: int) -> int:
        if inicio == fim:
            return lista_copia[inicio]

        pivot_index = particionar(inicio, fim)

        if k == pivot_index:
            return lista_copia[k]
        elif k < pivot_index:
            return k_menor(inicio, pivot_index - 1, k)
        else:
            return k_menor(pivot_index + 1, fim, k)

    meio = n // 2
    is_odd = bool(n & 1)

    if is_odd:
        return float(k_menor(0, n - 1, meio))
    primeiro, segundo = k_menor(0, n - 1, meio - 1), k_menor(0, n - 1, meio)
    return (primeiro + segundo) / 2

testes = [
    [1, 2, 3],
    [3, 2, 1],
    [1, 1, 2, 3],
    random.sample(range(-10, 10), 10),
    random.sample(range(-15, 15), 11)
]

for teste in testes:
    print(f'Lista: {teste}')
    print(f'Mediana: {mediana(teste)}')
```
