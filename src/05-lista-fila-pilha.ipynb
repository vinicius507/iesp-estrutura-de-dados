# Lista, Fila e Pilha

## Setup

```python
from typing import override

class Pilha[T]:
    def __init__(self) -> None:
        self.itens: list[T] = []

    def empilhar(self, item: T) -> None:
        """Insere um item no topo da lista."""

        self.itens.append(item)

    def desempilhar(self) -> T | None:
        """Remove o item no topo da pilha, se não estiver vazia."""

        if self.vazia():
            return None
        return self.itens.pop()

    def topo(self) -> T | None:
        """Retorna o item no topo da pilha sem remover."""

        if self.vazia():
            return None
        return self.itens[-1]

    def vazia(self) -> bool:
        """Verifica se a pilha está vazia."""

        return len(self.itens) == 0

    def __len__(self) -> int:
        return len(self.itens)

    @override
    def __str__(self) -> str:
        if self.vazia():
            return "Pilha vazia"

        itens = ", ".join((str(i) for i in self.itens))
        return f"Pilha: {itens}"

class Fila[T]:
    def __init__(self) -> None:
        self.itens: list[T] = []

    def enfileirar(self, item: T) -> None:
        """Adiciona um item na fila."""

        self.itens.append(item)

    def desenfileirar(self) -> T | None:
        """Remove o primeiro item da fila."""

        if self.vazia():
            return None
        return self.itens.pop(0)

    def vazia(self) -> bool:
        """Verifica se a fila está vazia."""

        return len(self.itens) == 0

    def primeiro(self) -> T | None:
        """Verifica o primeiro elemento da fila."""

        if self.vazia():
            return None
        return self.itens[0]

    def ultimo(self) -> T | None:
        """Verifica o ultimo elemento da fila."""

        if self.vazia():
            return None
        return self.itens[-1]

    def __len__(self) -> int:
        return len(self.itens)

    @override
    def __str__(self) -> str:
        if self.vazia():
            return "Fila vazia"

        itens = ", ".join((str(i) for i in self.itens))
        return f"Fila: {itens}"
```

1. Você está desenvolvendo um sistema de fila de impressão para uma empresa. Os documentos são adicionados à fila e impressos na ordem em que foram recebidos. Implemente um programa Python que use a classe de fila para simular esse processo. Apresente algumas situações para inserir documentos na fila e verificar situações como: o primeiro da fila, o último da fila, quantos estão na fila.

```python
class SistemaFilaImpressao:
    def __init__(self) -> None:
        self.fila: Fila[str] = Fila[str]()

    def adicionar_documento(self, documento: str) -> None:
        self.fila.enfileirar(documento)
        print(f"Documento '{documento}' adicionado à fila de impressão.")

    def imprimir_documento(self) -> None:
        documento = self.fila.desenfileirar()
        if documento is None:
            print("Nenhum documento para imprimir.")
            return
        print(f"Imprimindo: {documento}")

    def detalhes_fila(self) -> None:
        print(f"Fila de impressão: {self.fila}")
        print(f"Primeiro da fila: {self.fila.primeiro()}")
        print(f"Último da fila: {self.fila.ultimo()}")
        print(f"Quantidade de documentos na fila: {len(self.fila)}")

fila_impressao = SistemaFilaImpressao()
fila_impressao.detalhes_fila()
fila_impressao.adicionar_documento("CNH.pdf")
fila_impressao.adicionar_documento("Apresentação.pptx")
fila_impressao.detalhes_fila()
fila_impressao.imprimir_documento()
fila_impressao.detalhes_fila()
fila_impressao.imprimir_documento()
fila_impressao.adicionar_documento("README.md")
fila_impressao.adicionar_documento("LICENSE")
fila_impressao.detalhes_fila()
fila_impressao.imprimir_documento()
fila_impressao.imprimir_documento()
fila_impressao.imprimir_documento()
fila_impressao.imprimir_documento()
```

6. Imagine que você está desenvolvendo um navegador web simplificado. Use uma pilha para armazenar o histórico de páginas visitadas pelos usuários e implementar as funcionalidades de voltar e avançar na navegação.

```python
class SistemaHistorico:
    def __init__(self) -> None:
        self.pagina_atual: str | None = None
        self.pilha_voltar: Pilha[str] = Pilha[str]()
        self.pilha_avancar: Pilha[str] = Pilha[str]()

    def visitar_pagina(self, pagina: str) -> None:
        if self.pagina_atual is not None:
            self.pilha_voltar.empilhar(self.pagina_atual)
            self.pilha_avancar = Pilha[str]()  # Limpa a pilha de avançar
        self.pagina_atual = pagina
        print(f"Visitando: {pagina}")

    def voltar(self) -> None:
        if self.pilha_voltar.vazia():
            print("Não há páginas para voltar.")
            return

        if self.pagina_atual is not None:
            self.pilha_avancar.empilhar(self.pagina_atual)
        self.pagina_atual = self.pilha_voltar.desempilhar()
        print(f"Voltando para: {self.pagina_atual}")

    def avancar(self) -> None:
        if self.pilha_avancar.vazia():
            print("Não há páginas para avançar.")
            return
        if self.pagina_atual is not None:
            self.pilha_voltar.empilhar(self.pagina_atual)
        self.pagina_atual = self.pilha_avancar.desempilhar()
        print(f"Avançando para: {self.pagina_atual}")

    def detalhes_historico(self) -> None:
        print(f"Página atual: {self.pagina_atual}")
        print(f"Páginas para voltar: {self.pilha_voltar}")
        print(f"Páginas para avançar: {self.pilha_avancar}")

historico = SistemaHistorico()
historico.visitar_pagina("google.com")
historico.visitar_pagina("youtube.com")
historico.visitar_pagina("github.com")
historico.detalhes_historico()
historico.voltar()
historico.voltar()
historico.detalhes_historico()
historico.avancar()
historico.detalhes_historico()
historico.visitar_pagina("stackoverflow.com")
historico.detalhes_historico()
```

7. Crie uma calculadora que avalia expressões matemáticas no formato Notação Polonesa Reversa (RPN). Use uma pilha para armazenar os operandos e operadores e realizar os cálculos.

```python
import operator
from typing import Callable

OPERADORES: dict[str, Callable[[float, float], float]] = {
    "+": operator.add,
    "-": operator.sub,
    "*": operator.mul,
    "/": operator.truediv,
}

class CalculadoraRPN:
    def avaliar(self, expressao: str) -> float:
        pilha = Pilha[float]()

        for token in expressao.split():
            if token not in OPERADORES:
                pilha.empilhar(float(token))
                continue
            segundo = pilha.desempilhar()
            primeiro = pilha.desempilhar()

            if primeiro is None or segundo is None:
                raise ValueError("Expressão inválida")

            op = OPERADORES[token]
            resultado = op(primeiro, segundo)
            pilha.empilhar(resultado)

        resultado_final = pilha.desempilhar()

        if resultado_final is None or not pilha.vazia():
            raise ValueError("Expressão inválida")
        return resultado_final

testes = [
    ("3 4 +", 7.0),
    ("10 5 -", 5.0),
    ("2 3 *", 6.0),
    ("8 4 /", 2.0),
    ("3 4 + 2 *", 14.0),
    ("5 1 2 + 4 * + 3 -", 14.0),
]

calculadora = CalculadoraRPN()

for expressao, resultado_esperado in testes:
    resultado = calculadora.avaliar(expressao)

    if resultado == resultado_esperado:
        print(f"Teste passou: {expressao} = {resultado}")
        continue
    print(f"Teste falhou: {expressao} = {resultado}, esperado {resultado_esperado}")
```
